```python
import matplotlib.pyplot as plt
from runge_kutta_4 import runge_kutta_4
from math import pi, sin

def plot(table, title):
    ts = list(map(lambda res: res[0], table))
    xs = list(map(lambda res: deg(res[1]), table))
    ys = list(map(lambda res: deg(res[2]), table))

    plt.figure(figsize=(13, 7))
    plt.plot(ts, xs, label="Ángulo")
    plt.plot(ts, ys, label="Velocidad Angular")

    plt.title(title)
    plt.xlabel("t [s]")
    plt.ylabel("θ [°]     dθ/dt [°/s]")
    plt.legend()
    plt.show()

def rad(x):
    return x * pi / 180

def deg(x):
    return x * 180 / pi
```

# Sistema no amortiguado

```python
m = 1
L = 1
b = 0

def f(t, x, y):
    return y

def g(t, x, y):
    return - (b / m) * y - (9.81 / L) * sin(x)

table = runge_kutta_4(f, g, rad(30), rad(0), 0.2, 0, 20)
plot(table, "Sistema no amortiguado")
```

# Sistema con amortiguamiento subcrítico

```python
m = 1
L = 1
b = 0.5

def f(t, x, y):
    return y

def g(t, x, y):
    return - (b / m) * y - (9.81 / L) * sin(x)

table = runge_kutta_4(f, g, rad(30), rad(100), 0.2, 0, 20)
plot(table, "Sistema con amortiguamiento subcrítico")
```
